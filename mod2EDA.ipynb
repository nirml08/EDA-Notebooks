# Handling Missing values

```python
data.isnull().sum()
```

```python
# There are no missing values but there are some question marks which need to be handled
# replace ? with np.nan
data.replace('?',np.nan,inplace=True)
```

```python
data.isnull().sum()
```

 normalized-losses ,num-of-doors , bore , stroke , horsepower , peak-rpm,price





```python
# price
data['price']
```

```python
# dtype of price
data['price'].dtype
# since price values are integer but its datatype is object(wrong data type)
```

```python
#
plt.figure(figsize=(3,3))
sns.histplot(data['normalized-losses'].astype('float64'),bins=10)
```

```python
# clearly price distribution is skewed so will go with median
data['price'].median()
```

```python
data['price'].fillna(data['price'].median(),inplace=True)
```

# peak-rpm


```python
data['peak-rpm']
```

```python
data['peak-rpm'].dtype
```

```python
# since values within peak-rpm is integer but given object data type(wrong)
plt.figure(figsize=(3,3))
sns.histplot(data['peak-rpm'].astype('float64'),bins=10)
```

```python
# replace with median since distribution is skewed
data['peak-rpm'].median()
```

```python
data['peak-rpm'].fillna(data['peak-rpm'].median(),inplace=True)
```

```python
# bore , stroke and horsepower
data['bore'].fillna(data['bore'].median(),inplace=True)
data['stroke'].fillna(data['stroke'].median(),inplace=True)
data['horsepower'].fillna(data['horsepower'].median(),inplace=True)
data['normalized-losses'].fillna(data['normalized-losses'].median(),inplace=True)
```

# Exploratort data Analysis
* Univariate:histogram ,countplot , boxplot
* Bivariate : bar , line , scatter , pie , box/violin
* Multivariate: Pairplot , heat map

# Hints
Univariate
* Histogram : 1 numerical col---> to check distribution:Normal/skewed
* Count plot : 1 Categorical column ---> To check count of categories/value_count
* Box plot : 1 numerical column ---> Used to check outliers
--------------------
Bivariate
* Line : 1 numerical , 1 date column ---> ANalyse changes over period of time
* scatter plot: 2 numerical cols ---> find relationship between two variable
* bar plot: 1 numerical and categorical ---> analyse categories with numerical data
* pie chart : 1 num , 1 cat ----> proportion of whole
* Box plot/violin : 1 numerical and 1 categorical ---> compare distributions
-------------------
Multivariate
* pairplot : numerical data---> scatter plot for all numerical cols
* Heatmap : numerical data ----> correlation plot
